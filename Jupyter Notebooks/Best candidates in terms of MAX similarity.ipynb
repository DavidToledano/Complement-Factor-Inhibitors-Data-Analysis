```python
#Imports
import pandas as pd
```

```python
df = pd.read_excel('list_of_molecules.xlsx',sheet_name="Sheet1")
```

## 1/ Best candidates in terms of additive similarity - ECFP4 processing


Let's dive into this data File.
First, we will sum the similarity rates obtained with ECFP4 processing and identify the molecular candidates with the highest value

```python
df_ECFP4 = df[['COMPOUND_ID', 'Cluster', 'Rosmarinic_ClosestSimilarities_ECFP4', 'SB_ClosestSimilarities_ECFP4' ]]
```

```python
df_ECFP4['ECFP4_sum'] = df_ECFP4['Rosmarinic_ClosestSimilarities_ECFP4'] + df_ECFP4['SB_ClosestSimilarities_ECFP4']
```

```python
df_ECFP4 = df_ECFP4[~(df_ECFP4['Rosmarinic_ClosestSimilarities_ECFP4'].isnull() | df_ECFP4['SB_ClosestSimilarities_ECFP4'].isnull())]
```

```python
idx = df_ECFP4.groupby(['Cluster'])['ECFP4_sum'].transform(max) == df_ECFP4['ECFP4_sum']
```

```python
df_ECFP4_clust_cand = df_ECFP4[idx].sort_values(by=['Cluster'])
```

```python
df_ECFP4_clust_cand = df_ECFP4_clust_cand.reset_index()
```

### DataFrame1: Best candidates in terms of additive similarity - ECFP4 processing

```python
df_ECFP4_clust_cand
```

## 2/ Best candidates in terms of SB similarity - CREPO processing

Let's know get the maximum similarity with SB by CREPO processing 


```python
df_CREPO_SB = df[['COMPOUND_ID', 'Cluster', 'CREPO_SB_Similarity']]
```

```python
df_CREPO_SB = df_CREPO_SB[~(df_CREPO_SB['CREPO_SB_Similarity'].isnull())]
```

```python
idx = df_CREPO_SB.groupby(['Cluster'])['CREPO_SB_Similarity'].transform(max) == df_CREPO_SB['CREPO_SB_Similarity']
```

```python
df_CREPO_SB_clust_cand = df_CREPO_SB[idx].sort_values(by=['Cluster'])
```

### DataFrame2: Best candidates in terms of SB similarity - CREPO processing

```python
df_CREPO_SB_clust_cand
```

## 3/ Best candidates in terms of RA similarity - CREPO processing

```python
df_CREPO_RA = df[['COMPOUND_ID', 'Cluster', 'CREPO_Rosmarinic_Similarity']]
```

```python
df_CREPO_RA = df_CREPO_RA[~(df_CREPO_RA['CREPO_Rosmarinic_Similarity'].isnull())]
```

## 4/ Merging these 3 dataframe

```python
ID_ECFP4_clust_cand = df_ECFP4_clust_cand['COMPOUND_ID']
```

```python
ID_CREPO_RA_clust_cand = df_CREPO_RA_clust_cand['COMPOUND_ID']
```

```python
ID_CREPO_SB_clust_cand = df_CREPO_SB_clust_cand['COMPOUND_ID']
```

```python
ID_clus_cand = pd.concat([ID_ECFP4_clust_cand, ID_CREPO_RA_clust_cand, ID_CREPO_SB_clust_cand])
```

```python
ID_clus_cand.count()
```

We want to make sure that there are no duplicates, even if we know that there should because in this datafile, it either ECFP4 or CREPO never both

```python
ID_clus_cand.drop_duplicates()
```

```python
ID_clus_cand.count()
```

Indeed there are no duplicates

```python
df_clust_cand = df.loc[df['COMPOUND_ID'].isin(ID_clus_cand)]
```

```python
df_clust_cand = df_clust_cand.sort_values(by=['Cluster'])
```

```python
df_clust_cand = df_clust_cand.reset_index()
```

```python
first_column = df_clust_cand.pop('Cluster')
```

```python
df_clust_cand.insert(0, 'Cluster', first_column)
```
